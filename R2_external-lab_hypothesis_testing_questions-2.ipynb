```python
import pandas as pd
Experiment_df = pd.DataFrame()

df1            = pd.DataFrame({'Exp': '1', 'Experiment':e1})
df2            = pd.DataFrame({'Exp': '2', 'Experiment':e2})
df3            = pd.DataFrame({'Exp': '3', 'Experiment':e3})

Experiment_df = Experiment_df.append(df1) 
Experiment_df = Experiment_df.append(df2) 
Experiment_df = Experiment_df.append(df3) 
```

## Question 10

A Paired sample t-test compares means from the same group at different times.

The basic two sample t-test is designed for testing differences between independent groups. 
In some cases, you might be interested in testing differences between samples of the same group at different points in time. 
We can conduct a paired t-test using the scipy function stats.ttest_rel(). 

```python
before= stats.norm.rvs(scale=30, loc=100, size=500) ## Creates a normal distribution with a mean value of 100 and std of 30
after = before + stats.norm.rvs(scale=5, loc=-1.25, size=500)
```

**Step 1: Define null and alternative hypothesis**
    
In testing whether weight-loss drug has any effect on patients.

The null hypothesis states that difference in weight loss,ùúáùê¥ùëìùë°ùëíùëü equals ùúáùêµùëíùëìùëúùëüùëí

The alternate hypothesis states that difference in the weight loss,  ùúáùê¥ùëìùë°ùëíùëü‚â†ùúáùêµùëíùëìùëúùëüùëí

**Step 2: Decide the significance level**

Here we select Œ±= 0.05 and sample size < 500 

**Step 3: Identify the test statistics**
    
Sample sizes for both samples are same.
We have two paired samples

**Step 4: Calculate the p - value and test statistic**
    
We use the scipy.stats.ttest_rel to calculate the T-test on TWO RELATED samples of scores. This is a
two-sided test for the null hypothesis that 2 related or repeated samples have identical average (expected) values.
Here we give the two sample observations as input. This function returns t statistic and two-tailed p value.

```python
t_statistic, p_value  =  stats.ttest_rel(before,after)
print(p_value)
```

**Step 5: Decide to reject or accept null hypothesis**
    
In this example, p value is 9.931381206056462e-08 and it is less than 5% level of significance
So the statistical decision is to reject the null hypothesis at 5% level of significance.
