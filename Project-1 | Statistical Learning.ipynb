```python
##1-->Reading the Laliga CSV File

import pandas as pd
import numpy as np

dfliga=pd.read_csv("Laliga.csv",header=1)
dfliga.head(10)
```

```python
##1-->Replacing the '-' to 0 across the DataFrame

dfrep=dfliga.replace('-',0)
dfrep.head(10)
```

```python
##2-->Teams which has statrted playing between 1930-1980

df_debut = dfrep["Debut"].between('1930', '1980', inclusive = True)
dfdt=dfrep[df_debut]
dfdt[['Debut','Team']]
```

```python
##3-->Getting the Five top Teams from Lalliga Dataframe

df5t=dfrep[['Team','Points']]
df5t['Points']=df5t['Points'].astype('int')
df5t.sort_values(by='Points',ascending=False)
df5t.head(5)
```

```python
#4-->Goal Difference Count Function
#Output is a New column Goal difference count
#Maximun and Minimum Goal Difference count


def Goal_diff_count(x):
    dfread=pd.read_csv("Laliga.csv",header=1)
    dfcs=dfread[['Team','GoalsFor','GoalsAgainst']]
    dfrepl=dfcs.replace('-',0)
    dfrepl[['GoalsFor', 'GoalsAgainst']] = dfrepl[['GoalsFor', 'GoalsAgainst']].apply(pd.to_numeric) 
    dfti=dfrepl.set_index('Team')
    dfti['Goal_diff_count']= dfti['GoalsFor'] -dfti['GoalsAgainst']
    print(dfti)
    
    print(dfti['Goal_diff_count'].max())
    print(dfti['Goal_diff_count'].min())
Goal_diff_count('Laliga.csv')
```

```python
#5-->Create a new column with name “Winning Percent” and append it to the data set
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

dfrep['GamesWon'] =pd.to_numeric(dfrep['GamesWon'])
dfrep['GamesPlayed']=pd.to_numeric(dfrep['GamesPlayed'])
dfrep['Winning Percent']=dfrep['GamesWon']/dfrep['GamesPlayed'] *100
dfrep.dtypes

dfrep.sort_values(by='Winning Percent',ascending=False)
print(dfrep.iloc[0:,7:].fillna(0).head(5))
```
