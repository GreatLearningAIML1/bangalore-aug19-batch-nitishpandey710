## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class:

### Find out the mean and variance for each column but for class column 

```python
#Mean and Variance for Sepal Length
MeanSL=Iris['Sepal Length (in cm)'].mean()
VarSL=Iris['Sepal Length (in cm)'].var()

print('Sepal Length Mean :',MeanSL)
print('Sepal Length Variance :',VarSL)

#Mean and Variance for Sepal width
MeanSW=Iris['Sepal Width in (cm)'].mean()
VarSW=Iris['Sepal Width in (cm)'].var()

print ('Sepal Width Mean :',MeanSW)
print ('Sepal Width Variance :',VarSW)

#Mean and Variance for Petal Length
MeanPL=Iris['Petal length (in cm)'].mean()
VarPL=Iris['Petal width (in cm)'].var()

print ('Petal length Mean :',MeanPL)
print ('Petal length Variance :',VarPL)

#Mean and Variance for Petal Width
MeanPW=Iris['Petal width (in cm)'].mean()
VarPW=Iris['Petal width (in cm)'].var()

print ('Petal width Mean :',MeanPW)
print ('Petal width Variance :',VarPW)
```

### Find out the IQR for the column - Age

```python
#Quantile function is used to get the 75th Percentile
SeventyFifth=Hrdataset.Age.quantile(0.75)
print(SeventyFifth)

#Quantile function is used to get the 25th Percentile
TwentyFifth=Hrdataset.Age.quantile(0.25)
print(TwentyFifth)

#Getting Interquartile

Interquartile = SeventyFifth - TwentyFifth
print(Interquartile)
```
